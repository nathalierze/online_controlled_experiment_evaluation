```python
import pandas as pd
import pickle
```

```python
df = pd.read_csv( 'learning_data.csv')
df
```

```python
# user id as index
df=df.set_index('UserID')
# drop rows with 0 values
df = df.dropna(how='all')
# create index with four digits
df.index = df.index.astype(str).str.pad(7, side='left', fillchar='0')
# change 1.0 to 1
df = df.replace([1.0],'1')
# change 0.0 to 0
df = df.replace([0.0],'0')
# change NaN to .
df = df.fillna('.', inplace=False)
```

```python
df
```

```python
df.to_csv('conquest_input_data.csv')
```
